```python
import pandas as pd
```

```python
#import CSV files for RBs
rb2012_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2012.csv", error_bad_lines=False)
rb2013_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2013.csv", error_bad_lines=False)
rb2014_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2014.csv", error_bad_lines=False)
rb2015_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2015.csv", error_bad_lines=False)
rb2016_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2016.csv", error_bad_lines=False)
rb2017_data = pd.read_csv("RB_Data/fantasy-football-leadersRB2017.csv", error_bad_lines=False)
```

```python
#Combine the data frames for RBs
rb_frames = [rb2012_data, rb2013_data, rb2014_data, rb2015_data, rb2016_data, rb2017_data]
rb_result = pd.concat(rb_frames)
rb_result.head()
```

```python
# Put RB Data into new CSV
rb_result.to_csv('Totals_Data/RB_Total_data.csv')
```
