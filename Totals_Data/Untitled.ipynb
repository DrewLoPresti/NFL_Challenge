```python
import pandas as pd
```

```python
#Bring in Combine Data
Combine_Total_data = pd.read_csv("Combine_Total_data.csv", error_bad_lines=False)

# Bring in Snap Count Data
Sum_DB_Snap_Count = pd.read_csv("Sum_DB_Snap_Count.csv", error_bad_lines=False)
Sum_LB_Snap_Count = pd.read_csv("Sum_LB_Snap_Count.csv", error_bad_lines=False)
Sum_RB_Snap_Count = pd.read_csv("Sum_RB_Snap_Count.csv", error_bad_lines=False)
Sum_QB_Snap_Count = pd.read_csv("Sum_QB_Snap_Count.csv", error_bad_lines=False)
Sum_WR_Snap_Count = pd.read_csv("Sum_WR_Snap_Count.csv", error_bad_lines=False)
```

```python
#Combine the data frames for Combine and Snap Count
Combine_Total_data = Combine_Total_data.set_index("Name")
Sum_DB_Snap_Count = Sum_DB_Snap_Count.set_index("Name")
Sum_LB_Snap_Count = Sum_LB_Snap_Count.set_index("Name")
Sum_RB_Snap_Count = Sum_RB_Snap_Count.set_index("Name")
Sum_QB_Snap_Count = Sum_QB_Snap_Count.set_index("Name")
Sum_WR_Snap_Count = Sum_WR_Snap_Count.set_index("Name")
```

```python
Combine_position_frames = [Sum_DB_Snap_Count, Sum_LB_Snap_Count, Sum_RB_Snap_Count, Sum_QB_Snap_Count, Sum_WR_Snap_Count]
Combine_position_result = pd.concat(Combine_position_frames)
Combine_position_result

Merge = Combine_Total_data.merge(Combine_position_result, how="left", left_index = True, right_index = True)
Merge.head()
```

```python
#Combine_position_result.columns
Merge.to_csv("Trial")
```
