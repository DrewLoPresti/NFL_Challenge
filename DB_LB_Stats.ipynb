```python
import pandas as pd
```

```python
#DB Stats 2012 to 2017
db2012_data = pd.read_csv("DB_Data/fantasy-football-leaders 2012db.csv", error_bad_lines=False)
db2013_data = pd.read_csv("DB_Data/fantasy-football-leaders 2013db.csv", error_bad_lines=False)
db2014_data = pd.read_csv("DB_Data/fantasy-football-leaders 2014db.csv", error_bad_lines=False)
db2015_data = pd.read_csv("DB_Data/fantasy-football-leaders 2015db.csv", error_bad_lines=False)
db2016_data = pd.read_csv("DB_Data/fantasy-football-leaders 2016db.csv", error_bad_lines=False)
db2017_data = pd.read_csv("DB_Data/fantasy-football-leaders 2017db.csv", error_bad_lines=False)
```

```python
#LB Stats 2012 to 2017
lb2012_data = pd.read_csv("LB_Data/fantasy-football-leaders 2012lb.csv", error_bad_lines=False)
lb2013_data = pd.read_csv("LB_Data/fantasy-football-leaders 2013lb.csv", error_bad_lines=False)
lb2014_data = pd.read_csv("LB_Data/fantasy-football-leaders 2014lb.csv", error_bad_lines=False)
lb2015_data = pd.read_csv("LB_Data/fantasy-football-leaders 2015lb.csv", error_bad_lines=False)
lb2016_data = pd.read_csv("LB_Data/fantasy-football-leaders 2016lb.csv", error_bad_lines=False)
lb2017_data = pd.read_csv("LB_Data/fantasy-football-leaders 2017lb.csv", error_bad_lines=False)
```

```python
#Combine the data frames for DB
db_frames = [db2012_data, db2013_data, db2014_data, db2015_data, db2016_data, db2017_data]
db_result = pd.concat(db_frames)
db_result.head()
```

```python
# Put DB Data into new CSV
db_result.to_csv('Totals_Data/DB_Total_data.csv')
```

```python
# Combine the Data Frames for LB's
lb_frames = [lb2012_data, lb2013_data, lb2014_data, lb2015_data, lb2016_data, lb2017_data]
lb_result = pd.concat(lb_frames)
lb_result.head()
```

```python
# Put LB Data into new CSV
lb_result.to_csv('Totals_Data/LB_Total_data.csv')
```
