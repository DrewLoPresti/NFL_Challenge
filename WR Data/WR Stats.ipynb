```python
import pandas as pd
import numpy as np

wr_data = "../Totals_Data/WR_Total_data.csv"

wr = pd.read_csv(wr_data)
wr.head()
```

```python
wr.drop(['Unnamed: 0', 'Team', 'Position'], axis=1, inplace=True)
wr.head()
```

```python
wr.set_index('Name')
```

```python
wr_sum = wr.groupby(['Name']).sum()
```

```python
# Put Data into New CSV
wr_sum.to_csv('Sum_WR_Data.csv')
```
