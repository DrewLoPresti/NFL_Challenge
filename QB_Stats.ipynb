```python
import pandas as pd
import numpy
```

```python
#import CSV files for QBs
qb2012_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2012.csv", error_bad_lines=False)
qb2013_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2013.csv", error_bad_lines=False)
qb2014_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2014.csv", error_bad_lines=False)
qb2015_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2015.csv", error_bad_lines=False)
qb2016_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2016.csv", error_bad_lines=False)
qb2017_data = pd.read_csv("QB_Data/fantasy-football-leadersQB2017.csv", error_bad_lines=False)
```

```python
#Combine the data frames for QBs
qb_frames = [qb2012_data, qb2013_data, qb2014_data, qb2015_data, qb2016_data, qb2017_data]
qb_result = pd.concat(qb_frames)
qb_result.head()
```

```python
# Put Data into New CSV
qb_result.to_csv('Totals_Data/QB_Total_data.csv')
```

```python
mean_df = qb_result.drop(['Rank', 'Team', 'Position', 'PassingCompletions', 'PassingAttempts', 'PassingYards', 'PassingTouchdowns', 'PassingInterceptions', 'RushingAttempts', 'RushingYards', 'RushingTouchdowns','FantasyPointsPerGame', 'FantasyPoints'],axis=1)
sum_df = qb_result.drop(['Rank', 'Team', 'Position', 'PassingCompletionPercentage', 'PassingYardsPerAttempt', 'PassingRating', 'RushingYardsPerAttempt', 'FantasyPointsPerGame', 'FantasyPoints'],axis=1)
```

```python
mean_df_updated = mean_df.groupby(['Name']).mean()
sum_df_updated = sum_df.groupby(['Name']).sum()
```
